### Univariate Categorical

```python
def uni_cat(array1,array0):
    plt.figure(figsize = (20,5))
    
    plt.subplot(1,2,1)
    plt.title('Targeted')
    sns.countplot(array1.dropna())
    plt.plot()
    
    plt.subplot(1,2,2)
    plt.title('Non-Targeted')
    sns.countplot(array0.dropna())
    plt.plot()
    
def uni_cat_wrap(array1,array0):
    plt.figure(figsize = (20,5))
    
    plt.subplot(1,2,1)
    plt.title('Targeted')
    ax = sns.countplot(array1.dropna())
    ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
    plt.plot()
    
    plt.subplot(1,2,2)
    plt.title('Non-Targeted')
    ax = sns.countplot(array0.dropna())
    ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
    plt.plot()
    
```

```python
def heatmap_corr(array1,array2):
    plt.figure(figsize = (20,8))
    
    plt.subplot(1,2,1)
    plt.title('Targeted')
    sns.heatmap(array1.corr(), annot = True, cmap = 'Reds')
    plt.plot()
    
    plt.subplot(1,2,2)
    plt.title('Non-Targeted')
    sns.heatmap(array2.corr(), annot = True, cmap = 'Reds')
    plt.plot()
```

## Observations and Recommendations

### Defaulters:
    # For clients with amount credit of less than 1.5 million chances of defaulting are more, across all age groups
    # Clients with change in Id in last 15 years and loan amount less than 1.5 million are more likely to default
    # Unaccompanied clients are more likely to default
    # Single and Married people are more likely default
    # Home loans and car loans by clients are likely to default, hence proper check should be done
### Non Defaulters:
    # For clients with amount credit of more than 2 million chances of regular repayment are more, across all age groups
    # Clients with loan amount more than 1.5 millions are less likely to default
    # Clients who come in groups are more likely to repay
    # Widowed people generally pay loan amount on time
    # Clients with higher rate of interest are more likely to repay loan amount
    # If the rate of down payment is more than 60% than it does not matter when the client applied for previous loan. He is less likely to default
